Mengusir / menghapus outliers

```python
#ngitung ngene IQR 
def numerical_starts(df, columns, multiple=1.5):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.75) #quartil
        Q3 = df_clean[col].quantile(0.25) #quartil
        IQR = Q3 - Q1 #median
        lower_bound = Q1 - multiple * IQR
        upper_bound = Q3 - multiple * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col]) <= upper_bound]
    return df_clean

```

```python
df_no_outlayers = numerical_starts(dataFrame, numerical_cols, 3.0)
```

# Feature Engineering

Total purchases Num and Mnt

```python
dataFrame.info()
```

Column total pengeluaran dan pembelian

```python
dataFrame['Total_Purchases'] = (dataFrame['NumDealsPurchases'] + dataFrame['NumWebPurchases'] + dataFrame['NumCatalogPurchases'] + dataFrame['NumStorePurchases'])
```

```python
dataFrame['Total_Spending'] = (dataFrame['MntWines'] + dataFrame['MntMeatProducts'] + 
                              dataFrame['MntFishProducts'] + dataFrame['MntSweetProducts'] + 
                              dataFrame['MntGoldProds'])
```

```python
purchaces_channels = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
dataFrame['Channel_Diversity'] = (dataFrame[purchaces_channels] > 0).sum(axis=1)
dataFrame['Avg_Purchases_Value'] = dataFrame['Total_Spending'] / (dataFrame['Total_Purchases'] + 1)
```

Saya telah melakukan proses Feature Engineering pada dataset Marketing.csv guna menambah wawasan dalam analisis perilaku pelanggan. Dalam proses ini, saya membuat empat kolom baru, yaitu:

- Total_Purchases: Kolom ini dibuat untuk menghitung total pembelian produk oleh setiap pelanggan. Tujuannya adalah untuk mengetahui seberapa sering pelanggan melakukan transaksi.

- Total_Spending: Kolom ini menunjukkan total pengeluaran pelanggan, yakni akumulasi jumlah uang yang telah dibelanjakan untuk seluruh produk.

- Channel_Diversity: Kolom ini merepresentasikan keberagaman saluran pembelian yang digunakan pelanggan, seperti melalui website, katalog, maupun toko fisik. Ini membantu memahami preferensi saluran belanja pelanggan.

- Avg: Merupakan nilai rata-rata pengeluaran per pembelian yang dihitung dari rasio Total_Spending terhadap Total_Purchases. Ini memberikan gambaran tentang seberapa besar pengeluaran pelanggan setiap kali melakukan transaksi.

Tujuan dari pembuatan keempat fitur ini adalah untuk membantu dalam menganalisis pengeluaran dan pendapatan produk secara individual. Di antara keuntungannya :

- Mengidentifikasi pelanggan yang paling menguntungkan

- Memahami pola dan preferensi belanja mereka

- Menyusun strategi pemasaran yang lebih tepat sasaran

- Meningkatkan revenue dan mempertahankan loyalitas pelanggan (customer retention)
